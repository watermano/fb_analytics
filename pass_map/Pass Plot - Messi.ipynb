```python
import matplotlib.pyplot as plt
import numpy as np
from mplsoccer import Pitch, Sbopen, VerticalPitch
import seaborn as sns
import pandas as pd
from matplotlib.lines import Line2D
```

```python
df = pd.read_csv(r"C:\Users\daguajo\Downloads\messibetis.csv")
```

```python
#Convert the data

df['x'] = df['x']*1.2
df['y'] = df['y']*.8
df['endX'] = df['endX']*1.2
df['endY'] = df['endY']*.8
```

```python
#Create plot
fig,ax = plt.subplots(figsize = (13.5,8), constrained_layout=True, tight_layout=False)
fig.set_facecolor('#22312b')
ax.patch.set_facecolor('#22312b')

pitch = Pitch(pitch_type='statsbomb', pitch_color='#22312b', line_color='#c7d5cc')

pitch.draw(ax=ax)
plt.gca().invert_yaxis()

for x in range(len(df['x'])):
    if df['outcome'][x] == 'Successful':
        plt.plot((df['x'][x], df['endX'][x]),(df['y'][x], df['endY'][x]), color='green')
        plt.scatter(df['x'][x],df['y'][x], color='green')
        plt.arrow(df['x'][x],df['y'][x], df['endX'][x]-df['x'][x], df['endY'][x]-df['y'][x], length_includes_head=True,
    head_width=1, head_length=1.5,color='green')
    else:
        plt.plot((df['x'][x], df['endX'][x]),(df['y'][x], df['endY'][x]), color='red')
        plt.scatter(df['x'][x],df['y'][x], color='red')
        plt.arrow(df['x'][x],df['y'][x], df['endX'][x]-df['x'][x], df['endY'][x]-df['y'][x], length_includes_head=True,
    head_width=1, head_length=1.5,color='red')
        
legend_elements = [
    Line2D([0], [0], color='green', lw=3, label='Successful Pass'),
    Line2D([0], [0], color='red', lw=3, label='Unsuccessful Pass')
]

# Add legend
ax.legend(handles=legend_elements, facecolor='#22312b', edgecolor='None',
          fontsize=18, loc='upper left', labelcolor='white')
        
#Set title        
plt.title('Messi Pass Map vs Real Betis', color='white', size=20)

plt.show()
```
